### Składowe spójne

```python
g.graph = nx.relabel_nodes(g.graph, lambda x: x.replace('$', 'S')) # prblem ze znakiem $ ze względu na wywietlanie nazw przy użyciu latexa 
connected_components = list(nx.connected_components(g.graph))
```

```python
def show_subgraphs(list_nodes, title):
    subgraphs = [g.graph.subgraph(c).copy() for c in list_nodes if len(c) > 1]

    for i, subgraph in enumerate(subgraphs):
        fig, ax = plt.subplots(figsize=(5, 5))
        pos = nx.spring_layout(subgraph)
        nx.draw(subgraph, pos, with_labels=True, node_size=400)
        ax.set_title(f'{title} {i + 1}')

    plt.show()
```

```python
show_subgraphs(connected_components, 'Składowa')
```

## Kliki

```python
cliques = list(nx.find_cliques(g.graph)) # to znajduje max kliki
```

```python
show_subgraphs(cliques, 'Klika')
```

```python
largest_clique = max(nx.find_cliques(g.graph), key=len)
print(largest_clique)
print(f'Liczba wierzchołków : {len(largest_clique)}')
show_subgraphs([largest_clique], 'Największa klika')
```
